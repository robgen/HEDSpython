Since we have two dataframes, we want to merge by a common column. Since observations are reported by state, we can merge the two dataframes by the state column which they have in common.

However, if we first look at the shape of the dataframes we can see that df_food has 98 rows while df_gdp has 34 rows. We want one entry per state so we need to check that this is the case for df_gdp and the df_food.

```python
df_gdp.shape
```

```python
df_food.shape
```

If we look at df_food, we cans see that it reports observations for three different periods (starting 2005, 2015 and 2019). We select only observation reported from 2019.

```python
df_food = df_food[df_food["reference_period_start"] == "2019-01-01"]
```

```python
df_food
```

We check how many unique values are in the State/Uts column of the df_gdp dataframe and if this value is the same as the number of rows.

```python
states_gdp = df_gdp["State/Uts"].unique().tolist()
len(states_gdp)
```

It is!

We check the same for the provider_admin1_name column in the df_food dataframe

```python
states_food = df_food["provider_admin1_name"].unique().tolist()
len(states_food)
```

This isn't. There are a few more unique values in this case so we see if there may be any odd ones.

With the code below we check which values are in one unique state list and not the other.

```python
# Convert lists to sets
set_food = set(states_food)
set_gdp = set(states_gdp)

# Find elements in states_food but not in states_gdp
food_not_in_gdp = list(set_food - set_gdp)

# Find elements in states_gdp but not in states_food
gdp_not_in_food = list(set_gdp - set_food)

# Print the results
print("States in food data but not in GDP data:", food_not_in_gdp)
print("States in GDP data but not in food data:", gdp_not_in_food)
```

First, drop na values found in the 'provider_admin1_name' column.

```python
df_food = df_food.dropna(subset=['provider_admin1_name'])
```

Then, we choose a common denomination for Jammu & Kashmir.

```python
df_gdp[(df_gdp["State/Uts"] == "Jammu & Kashmir*") | (df_gdp["State/Uts"] == "Jammu & Kashmir-U.T.")]
```

It looks like `Jammu & Kashmir*` contains some data while `Jammu & Kashmir-U.T.` does not so we rename `Jammu & Kashmir*` and drop `Jammu & Kashmir-U.T.`.

```python
df_gdp.loc[df_gdp["State/Uts"] == "Jammu & Kashmir*", "State/Uts"] = "Jammu & Kashmir"
```

```python
df_gdp = df_gdp[~(df_gdp["State/Uts"] == "Jammu & Kashmir-U.T.")]
```

```python
df_gdp
```

We also drop the rows for 'Ladakh', 'Dadra & Nagar Haveli and Daman & Diu' and 'Lakshadweep' which are only found in df_food but not in df_gdp.

```python
df_food = df_food[~df_food['provider_admin1_name'].isin(['Ladakh', 'Dadra & Nagar Haveli and Daman & Diu', 'Lakshadweep'])]
```

Now that we two clean datasets reporting observations for the same states we can merge the dataframes by the state columns.

```python
df = pd.merge(df_food, df_gdp, left_on="provider_admin1_name", right_on="State/Uts")
```
